```python
import os
import cv2

# Define dataset paths
lsa_path = r"datasets\lsa64_cut"
mendeley_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\processed_videos"

output_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\extracted_frames"

# Ensure output folder exists
os.makedirs(output_path, exist_ok=True)

# Function to extract frames from videos
def extract_frames(video_path, save_dir, frame_interval=5):
    cap = cv2.VideoCapture(video_path)
    count = 0
    frame_count = 0

    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break

        # Save every 'frame_interval' frame
        if count % frame_interval == 0:
            frame_filename = os.path.join(save_dir, f"frame_{frame_count}.jpg")
            cv2.imwrite(frame_filename, frame)
            frame_count += 1

        count += 1

    cap.release()

# Process LSA64 videos
for video_file in os.listdir(lsa_path):
    if video_file.endswith(".mp4"):
        video_class = video_file.split("_")[0]  # Extract label
        save_dir = os.path.join(output_path, "LSA64", video_class)
        os.makedirs(save_dir, exist_ok=True)
        extract_frames(os.path.join(lsa_path, video_file), save_dir)

# Process Mendeley videos
for video_file in os.listdir(mendeley_path):
    if video_file.endswith(".mov"):
        video_class = video_file.split("_")[0]  # Extract label
        save_dir = os.path.join(output_path, "Mendeley", video_class)
        os.makedirs(save_dir, exist_ok=True)
        extract_frames(os.path.join(mendeley_path, video_file), save_dir)

print("Frame extraction complete.")
```

```python
import os
import shutil

# Paths to datasets
asl_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\asl\asl_alphabet_train"
isl_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\Indian Sign Language Video and Text dataset for sentences (ISLVT)"
lsa_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\lsa64_cut"
indian_path= r"D:\ML PROJECTS\AI Sign Language Translator\datasets\indian"

# Define target paths
image_dataset_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\merged_dataset\images"
video_dataset_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\merged_dataset\videos"

# Ensure merged dataset folders exist
os.makedirs(image_dataset_path, exist_ok=True)
os.makedirs(video_dataset_path, exist_ok=True)

# Merge function
def merge_datasets(src_folder, img_dest_folder, vid_dest_folder):
    if not os.path.exists(src_folder):
        print(f"⚠️ Warning: {src_folder} does not exist. Skipping...")
        return

    for item in os.listdir(src_folder):
        src_path = os.path.join(src_folder, item)

        if os.path.isdir(src_path):  # If it's a folder (A-Z, categories)
            dest_label_path = os.path.join(img_dest_folder, item.lower())
            os.makedirs(dest_label_path, exist_ok=True)
            for file in os.listdir(src_path):
                file_path = os.path.join(src_path, file)
                if file.lower().endswith(('.png', '.jpg', '.jpeg')):  # Only images go here
                    shutil.copy(file_path, os.path.join(dest_label_path, file))
        
        elif os.path.isfile(src_path):  # Direct files (like .mp4 videos)
            if src_path.lower().endswith(('.mp4', '.mov', '.avi')):  # Only videos go here
                shutil.copy(src_path, os.path.join(vid_dest_folder, item))

# Merge ASL, ISL, LSA64 into respective folders
merge_datasets(asl_path, image_dataset_path, video_dataset_path)
merge_datasets(isl_path, image_dataset_path, video_dataset_path)
merge_datasets(lsa_path, image_dataset_path, video_dataset_path)
merge_datasets(indian_path, image_dataset_path, video_dataset_path)

print("✅ Static Dataset Merging Done!")

```

```python
import os
import cv2
import numpy as np
from tqdm import tqdm

# Paths
image_dataset_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\merged_dataset\images"
preprocessed_image_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\preprocessed\images"

# Ensure preprocessed folder exists
os.makedirs(preprocessed_image_path, exist_ok=True)

# Image processing function
def preprocess_images(src_folder, dest_folder, img_size=(224, 224)):
    for label in tqdm(os.listdir(src_folder), desc="Processing Images"):
        label_path = os.path.join(src_folder, label)
        dest_label_path = os.path.join(dest_folder, label)
        os.makedirs(dest_label_path, exist_ok=True)

        for file in os.listdir(label_path):
            file_path = os.path.join(label_path, file)
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                img = cv2.imread(file_path)
                img = cv2.resize(img, img_size)  # Resize to 224x224
                img = img / 255.0  # Normalize pixel values
                save_path = os.path.join(dest_label_path, file)
                cv2.imwrite(save_path, img * 255)  # Save back as an image

# Run preprocessing
preprocess_images(image_dataset_path, preprocessed_image_path)
print("✅ Image Preprocessing Done!")
```

```python
import os
import cv2
import numpy as np
from tqdm import tqdm

# Paths
video_dataset_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\merged_dataset\videos"
preprocessed_video_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\preprocessed\videos"

# Ensure output directory exists
os.makedirs(preprocessed_video_path, exist_ok=True)

# Video Frame Extraction Function
def extract_frames(video_folder, output_folder, frame_rate=5):
    if not os.path.exists(video_folder):
        print(f"⚠️ Warning: {video_folder} does not exist. Skipping video processing...")
        return

    for video in tqdm(os.listdir(video_folder), desc="Processing Videos"):
        video_path = os.path.join(video_folder, video)
        video_name, ext = os.path.splitext(video)

        if ext.lower() not in ['.mp4', '.mov', '.avi']:
            print(f"⚠️ Skipping non-video file: {video}")
            continue

        output_video_path = os.path.join(output_folder, video_name)
        os.makedirs(output_video_path, exist_ok=True)

        try:
            cap = cv2.VideoCapture(video_path)
            fps = cap.get(cv2.CAP_PROP_FPS) or 30  # Default FPS to 30 if undetected
            frame_interval = max(1, int(fps / frame_rate))  # Extract every Nth frame
            frame_count = 0

            while cap.isOpened():
                ret, frame = cap.read()
                if not ret:
                    break

                if frame_count % frame_interval == 0:
                    frame_filename = os.path.join(output_video_path, f"frame_{frame_count}.jpg")
                    cv2.imwrite(frame_filename, frame)

                frame_count += 1

            cap.release()
        except Exception as e:
            print(f"❌ Error processing {video}: {e}")

# Run Preprocessing
extract_frames(video_dataset_path, preprocessed_video_path)

print("✅ Video Preprocessing Done!")

```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Paths
train_data_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\preprocessed\images"

# Data Augmentation & Preprocessing
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = datagen.flow_from_directory(
    train_data_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='training'
)

val_generator = datagen.flow_from_directory(
    train_data_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    subset='validation'
)

# Define CNN Model
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(len(train_generator.class_indices), activation='softmax')
])

# Compile Model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train Model
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=5
)

# Save Model
model.save("sign_language_cnn.h5")

print("✅ CNN Training Completed & Model Saved!")
```

```python
import os
import numpy as np
import h5py
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Path to ASL Alphabet Train dataset
asl_train_path = r"D:\ML PROJECTS\AI Sign Language Translator\datasets\asl\asl_alphabet_train"
image_size = (64, 64)  # Resize all images to 64x64 pixels

# Define ImageDataGenerator for batch processing
datagen = ImageDataGenerator(rescale=1./255)

def preprocess_asl_images_with_gen(dataset_path, image_size, batch_size=32, save_path_images='asl_images.h5', save_path_labels='asl_labels.h5'):
    # Create a generator for the entire directory (this automatically handles the folder structure)
    generator = datagen.flow_from_directory(
        dataset_path,
        target_size=image_size,
        class_mode='sparse',  # Use sparse labels (integer encoding)
        batch_size=batch_size,
        shuffle=True
    )
    
    # Initialize HDF5 file for saving images and labels
    with h5py.File(save_path_images, 'w') as hf_images, h5py.File(save_path_labels, 'w') as hf_labels:
        batch_count = 0
        
        # Create datasets to store the images and labels
        dataset_images = hf_images.create_dataset('images', (0, *image_size, 3), maxshape=(None, *image_size, 3), chunks=True)
        dataset_labels = hf_labels.create_dataset('labels', (0,), maxshape=(None,), chunks=True)
        
        for batch_images, batch_labels in generator:
            batch_count += 1
            
            # Resize the datasets if necessary
            dataset_images.resize((dataset_images.shape[0] + batch_images.shape[0]), axis=0)
            dataset_labels.resize((dataset_labels.shape[0] + batch_labels.shape[0]), axis=0)
            
            # Append the current batch to the datasets
            dataset_images[-batch_images.shape[0]:] = batch_images
            dataset_labels[-batch_labels.shape[0]:] = batch_labels
            
            if batch_count * batch_size >= 87000:  # Stop once we have processed all images
                break
    
    print("Preprocessing complete. Images and labels saved to HDF5 files.")
    return save_path_images, save_path_labels

# Preprocess images with batch processing
save_path_images, save_path_labels = preprocess_asl_images_with_gen(asl_train_path, image_size)

# The processed data is now stored in HDF5 files ('asl_images.h5', 'asl_labels.h5')
```

```python
import h5py
import numpy as np

# Load the preprocessed data
def load_data(image_path=r'D:\ML PROJECTS\AI Sign Language Translator\asl_images.h5', label_path=r'D:\ML PROJECTS\AI Sign Language Translator\asl_labels.h5'):
    with h5py.File(image_path, 'r') as hf_images, h5py.File(label_path, 'r') as hf_labels:
        images = hf_images['images'][:]  # Load images
        labels = hf_labels['labels'][:]  # Load labels
    return images, labels

asl_images, asl_labels = load_data()

# Print shape of loaded data
print(f"Images shape: {asl_images.shape}")  # Expected: (87000, 64, 64, 3)
print(f"Labels shape: {asl_labels.shape}")  # Expected: (87000,)
```

```python
from sklearn.model_selection import train_test_split

# Split into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(asl_images, asl_labels, test_size=0.2, random_state=42)

# Check shapes
print(f"Train Images: {X_train.shape}, Train Labels: {y_train.shape}")
print(f"Test Images: {X_test.shape}, Test Labels: {y_test.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50V2
from tensorflow.keras.layers import Dense, Flatten, Dropout, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Load ResNet-50V2 and reduce complexity to mimic ResNet-18
base_model = ResNet50V2(weights="imagenet", include_top=False, input_shape=(64, 64, 3))

# Freeze initial layers to use pre-trained weights
for layer in base_model.layers[:-30]:  # Fine-tune last 30 layers
    layer.trainable = False

# Add custom classifier head
x = GlobalAveragePooling2D()(base_model.output)  # Corrected
x = Dense(256, activation="relu")(x)
x = Dropout(0.4)(x)
x = Dense(128, activation="relu")(x)
x = Dropout(0.3)(x)
output_layer = Dense(29, activation="softmax")(x)  # Corrected

# Create final model
model = Model(inputs=base_model.input, outputs=output_layer)  # Now correctly linked

# Compile model
model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

# Model summary
model.summary()

```

```python
import signal
import sys
from tensorflow.keras.callbacks import Callback

# Define a global flag to stop training
stop_training = False

# Function to handle keyboard interrupt (Ctrl+C)
def signal_handler(sig, frame):
    global stop_training
    print("\nTraining interrupted by user! Stopping gracefully...")
    stop_training = True

# Register the signal handler
signal.signal(signal.SIGINT, signal_handler)

# Custom callback to check the stop flag
class StopTrainingCallback(Callback):
    def on_batch_end(self, batch, logs=None):
        if stop_training:
            self.model.stop_training = True

# Train the model with the custom stop callback
history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=20,  # Upper limit, can stop anytime
    batch_size=32,  # Adjust based on GPU
    verbose=1,
    callbacks=[StopTrainingCallback()]  # Add manual stop callback
)
```

```python
import os
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image

def load_model(model_path):
    return tf.keras.models.load_model(model_path)

def predict_letter(model, img_path, class_names):
    img = image.load_img(img_path, target_size=(64, 64))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0  # Normalize
    prediction = model.predict(img_array)
    return class_names[np.argmax(prediction)]

def main():
    model_path = input("Enter the path to the trained model: ")
    model = load_model(model_path)
    class_names = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ") + ["nothing", "space"]
    
    final_word = ""
    while True:
        img_path = input("Enter the path of the image to translate: ")
        if not os.path.exists(img_path):
            print("Invalid path. Try again.")
            continue
        
        letter = predict_letter(model, img_path, class_names)
        if letter == "nothing":
            letter = ""
        elif letter == "space":
            letter = " "
        final_word += letter
        
        more = input("Do you want to upload another image? (yes/no): ").strip().lower()
        if more != "yes":
            break
    
    print("Final Translated Word:", final_word)

if __name__ == "__main__":
    main()

```
